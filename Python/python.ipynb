*Question 1:* <br>
Write a program that takes a string as input, and counts the frequency of each word in the string, there might be repeated characters in the string. Your task is to find the highest frequency<br> and returns the length of the highest-frequency word.

```python
def count_frequency_word(s:str):
    list_of_words = s.split()
    freq = {}
    for i in list_of_words:
        if i in freq:
            freq[i] += 1
        else:
            freq[i] = 1
    key = list(freq.keys())
    values = list(freq.values())
    return len(key[values.index(max(values))])
    
```

```python
a = 'hello hello hello hey hey yo'
b = 'yes yes no no good good bad bad bad'

print('a ', count_frequency_word(a))
print('b ',count_frequency_word(b))


```

*Question 2:* <br> 
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if he can remove just one character at the index in the string, and the <br>remaining characters will occur the same number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

```python
def val_str(s:str):
    char_freq = {}
    for i in s :
        if i in char_freq:
            char_freq[i] += 1
        else:
            char_freq[i] = 1
    
    freq_val_count= {}
    for j in char_freq.values():
        if j in freq_val_count:
            freq_val_count[j] += 1
        else:
            freq_val_count[j] = 1
    
    if len(freq_val_count) == 1:
        return "YES"
    
    if len(freq_val_count)>2:
        return 'NO'
    
    freq1,count1 = freq_val_count.popitem()
    freq2,count2 = freq_val_count.popitem()

    if (freq1 == 1 and count1 == 1) or (freq2 == 1 and count2 == 1) or abs(freq1 - freq2) == 1:
        return "YES"
    else:
        return "NO"
```

```python
print(val_str('aabbccc'))
print(val_str('aabbbbb'))
```

```python
def download_data_to_excel(link:str):
    import wget 
    import json
    import pandas as pd
    import os
    #downloading the data
    wget.download(link)
    #readind the data
    data_name = link.split("/")[-1]
    filepath = f'./{data_name}'
    data_name = data_name.split('.')[0]
    with open(filepath,'r') as f:
        #loading the json file 
        #and selecting the pokemon id key
        data = json.load(f)['pokemon']
    df = pd.DataFrame(data)
    df.to_excel(f'{data_name}.xlsx',index=False)
    return os.path.join('.',f'{data_name}.xlsx')
```

```python
import requests
import json
from datetime import datetime
from bs4 import BeautifulSoup

# API URL
api_url = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Sending GET request to the API endpoint
response = requests.get(api_url)

# CHecking if responce is 200
if response.status_code == 200:
    # Parse the response JSON data
    data = json.loads(response.text)

    # Extracting desired information from the data
    id = int(data['id'])
    url = str(data['url'])
    show_name = str(data["name"])
    #sesson = int(data['sesson'])
    #number = int(data['number'])
    type = str(data['type'])
    airdate = data['premiered']
    airdata = datetime.fromisoformat(airdate)
    airtime = data['schedule']['time']
    airtime = datetime.strptime(airtime, '%H:%M')
    time_12hr = airtime.strftime('%I:%M %p')
    runtime = float(data['runtime'])
    average_rating = float(data['rating']['average'])
    summary = data['summary']
    soup = BeautifulSoup(summary,'html.parser')
    summary = str(soup.get_text())
    medium = str(data['image']['medium'])
    original = str(data['image']['original'])

    # Format and print the extracted information
    print('id: ',id)
    print('url: ',url)
    print('Show Name: ',show_name)
    #print('Sesson: ',sesson)
    #print('Number: ',number)
    print('Type: ',type)
    print('Airdate: ',airdata)
    print('Airtime: ',time_12hr)
    print('Runtime: ',runtime)
    print('Avg Rating: ',average_rating)
    print('Summary: ',summary)
    print('Medium: ',medium)
    print('Original: ',original)

else:
    # Display an error message if the request was not successful
    print("Error occurred while accessing the API.")
```

**Question 10 -** <br> Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import defaultdict

def count_pos_tags(text):
    # Tokenize the text into words
    tokens = word_tokenize(text)

    # Perform part-of-speech tagging
    tagged_tokens = pos_tag(tokens)

    # Initialize counters for different parts of speech
    pos_counts = defaultdict(int)

    # Count the number of verbs, nouns, pronouns, and adjectives
    for token, tag in tagged_tokens:
        if tag.startswith('V'):  # Verb
            pos_counts['verbs'] += 1
        elif tag.startswith('N'):  # Noun
            pos_counts['nouns'] += 1
        elif tag.startswith('PRP'):  # Pronoun
            pos_counts['pronouns'] += 1
        elif tag.startswith('JJ'):  # Adjective
            pos_counts['adjectives'] += 1

    return pos_counts

phrase = "The quick brown fox jumps over the lazy dog."
phrase2 = 'Hello Are you GOOD'
counts = count_pos_tags(phrase)
count2 = count_pos_tags(phrase2)

print(counts)
print(count2)
```
